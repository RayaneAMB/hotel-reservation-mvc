```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Clean the Data
# We are dropping the rows with missing (NaN) values to make our data accurate
df_clean = df.dropna().copy()
print(f"🧹 Data Cleaned! Original rows: {len(df)} -> Cleaned rows: {len(df_clean)}\n")

# 2. Set up the visual style
sns.set_theme(style="darkgrid")
plt.figure(figsize=(16, 6))

# --- Chart 1: Total Sales by Category (Bar Chart) ---
plt.subplot(1, 2, 1)
category_sales = df_clean.groupby('Category')['Total_Sales'].sum().reset_index()
category_sales = category_sales.sort_values(by='Total_Sales', ascending=False)

sns.barplot(data=category_sales, x='Category', y='Total_Sales', palette='magma')
plt.title('Total Sales by Product Category', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)

# --- Chart 2: Revenue Distribution by Region (Pie Chart) ---
plt.subplot(1, 2, 2)
region_sales = df_clean.groupby('Region')['Total_Sales'].sum()

plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette('pastel'))
plt.title('Revenue Distribution by Region', fontsize=16, fontweight='bold')

# Show the charts!
plt.tight_layout()
plt.show()
```
